```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# Path to your test image
image_path = "/content/temp4.jpeg"  # Replace with your image path

# Analyze age, gender, and emotion
result = DeepFace.analyze(
    img_path=image_path,
    actions=["age", "gender", "emotion"],
    enforce_detection=False,  # Skip if no face detected
)

# Display the image with predictions
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# Print results
print("Predicted Age:", result[0]["age"])
print("Predicted Gender:", result[0]["dominant_gender"])
print("Predicted Emotion:", result[0]["dominant_emotion"])

# Display the image with annotations
plt.imshow(image)
plt.axis("off")
plt.title(f"Age: {result[0]['age']}, Gender: {result[0]['dominant_gender']}, Emotion: {result[0]['dominant_emotion']}")
plt.show()
```
